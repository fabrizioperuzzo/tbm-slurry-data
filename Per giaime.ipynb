```python
import os
import shutil

path = os.getcwd()+'\\00'

files = []
filespath = []
# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for file in f:
            list = [int(s) for s in r.split('-') if s.isdigit()]
            prefix = list[0]
            
            fulloriginal = os.path.join(r, file)
            
            dest_dir = os.getcwd()+'\\02'       
            newname = str(prefix) + '_'+file
            fullnamefile = os.path.join(dest_dir, newname)
            
            os.rename(fulloriginal, fullnamefile)#rename
```
